```python
import oauth2 as oauth
import json
```

```python
import os
keyPath=os.path.join(os.getcwd(), 'src', 'twitter.properties')

f=open(keyPath,'r')
lines=f.readlines()

def getKey(keyPath):
    d=dict()
    f=open(keyPath,'r')
    lines=f.readlines()
    for line in lines:
        row=line.split('=')
        row0=row[0]
        d[row0]=row[1].strip()
    return d

key=getKey(keyPath)
```

```python
import twitter

auth = twitter.oauth.OAuth(key['ACCESSTOKEN'],key['ACCESSTOKENSECRET'],
                            key['CONSUMERKEY'], key['CONSUMERSECRET'])
_client = twitter.Twitter(auth=auth)
print _client
```

```python
_client.statuses.update(status="Hello Twitter 1 161130")
```
